# *_Matplotlib_* 

```python
from matplotlib import pyplot as plt
```

## Les Bases

Tracer simplement des courbes ...

- `plt.plot(x_values (–), y_values (|))` va créer un graph avec les valeurs voulues 
- `plt.show()` affichera le graph précédement créé avec plt.plot


###### Exemple:

```python
mois =list(range(1, 13))
depenses =[200, 120, 120, 300, 240, 220, 240, 300, 310, 330, 330, 320]
revenue = [100, 150, 180, 300, 150, 160, 140, 170, 150, 130, 140, 140]


plt.plot(mois, depenses)#(x —, y |) 
plt.plot(mois, revenue)
plt.show()
```

### Linestyles

Ajout de couleurs et style aux courbes

- `linestyle = '--'` Pour changer le style de la ligne, ici en pointiller  
- `color = 'purple'` choisir la couleur, ce référer aux "HTML Color Names" pour trouver les noms des différentes couleurs   
- `marker = 's'` Poser des marqueurs sur chaques points de la courbe

```python
plt.plot(mois, depenses, linestyle='--', color='purple')#(x |, y –) 
plt.plot(mois, revenue, color='#82edc9', marker='s')
plt.show()
```

### Zoomer avec `plt.axis()`

 Il peut-être utile de zoomer sur certaines parties d'un graphiques  
On peut utiliser pour cela `plt.axis()`, en lui passant une `list` des `min` et `max` des `x`et des `y` à afficher

- `plt.axis([min(x), max(x), min(y), max(y)])`


```python
plt.plot(mois, depenses, linestyle='--', color='purple')#(x |, y –) 
plt.plot(mois, revenue, color='#82edc9', marker='s')

plt.axis([2, 6, 130, 320])
#plt.axis([min(x), max(x), min(y), max(y)])

plt.show()
```

### Titres et noms des labels

- `plt.xlabel('..')` Titre vertical de l'axe des ordonnées `y` sur la gauche du plot  
- `plt.ylabel('...')` titre horizotal de l'axe des absisses `x` en dessous du plot    
- `plt.title('..')` titre du graphique ce place en haut


```python
plt.plot(mois, depenses, linestyle='--', color='red', marker='o')#(x |, y –) 

plt.ylabel("entrées / sortie d'argent")
plt.xlabel("temps en mois")
plt.title("visualisation de rentabilité")
```

### subplots   
pour implémenter plusieurs graphiques à la suite on utilise la commande qui suit.  
On lui passe en paramètres, le nombre de lignes puis de colonnes, comme pour créer un tableau de plusieurs graph, puis la position du graph dans se tableau 

- `plt.subplot(row, col, position)`  lignes, colonnes, et position du plot là dedans
- `plt.subplot_adjust()` permet d'ajuster la distance entre les plot pou pas qu'ils ne ce superpose.
          
        
        
        Les arguments possible de subplot_adjust() :
        - `left` `=` par défaut à 0.125, augmente ou diminue l'espace sur la gauche
        - `right` par défaut à 0.9
        - `bottom` marge en bas du plot, par défaut à 0.1
        - `top` marge au-dessus du plot à 0.1
        - `wspace` espace horizontale entre 2 plot les uns au-dessus des autres, à 0.2
        - `hspace` espace verticale enre 2 graph adjacents, à 0.2 par défaut
    

######  exemple

```python
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

# Subplot 1
plt.subplot(2, 1, 1)
plt.plot(x, straight_line)

# Subplot 2
plt.subplot(2, 2, 3)
plt.plot(x, parabola)

# Subplot 3
plt.subplot(2, 2, 4)
plt.plot(x, cubic)

plt.subplots_adjust(wspace=0.35, bottom=0.2)

plt.show()
```

```python
plt.title('I')
plt.subplot(2, 3, 1)
plt.plot(x, cubic)

plt.title('II')
plt.subplot(2, 3, 2)

plt.title('III')
plt.subplot(2, 3, 3)

plt.title('VI')
plt.subplot(2, 3, 4)
# 2 lignes, 3 colonnes, position 4
# se plot sera donc le 1er sur la 2ème ligne de la 1ère colonne

plt.title('V')
plt.subplot(2, 3, 5)

plt.title('VI')
plt.subplot(2, 3, 6)

plt.subplots_adjust(wspace=0.50, bottom=0.01, hspace=0.3)
plt.show()
```
