```python
from matplotlib import pyplot as plt
```

## Les Bases

Tracer simplement des courbes, donner une taille au graph, les afficher et les sauvegarder...

- `plt.plot(x_values (–), y_values (|))` va créer un graph avec les valeurs voulues 
- `plt.figure(figsize=(lg –, H |))` donnera la taille voulue au plot, argument `figsize=(lo,H)` prend un `tuple` 
- `figure.get_size_inches()` pour connaitre la taille d'une figure
- `plt.show()` affichera le graph précédement créé avec `plt.plot()`
- `plt.savefig('nom.png')` pour enregistrer un plot, une figure, avec un nom et sous format `.png`

###### Exemple:

```python
mois =list(range(1, 13))
depenses =[200, 120, 120, 300, 240, 220, 240, 300, 310, 330, 330, 320]
revenue = [100, 150, 180, 300, 150, 160, 140, 170, 150, 130, 140, 140]

plt.figure(figsize=(10, 3)) # la taille du plot (longueur, hauteur)

plt.plot(mois, depenses)#(x —, y |) 
plt.plot(mois, revenue)

plt.show()
```

### Linestyles

Ajout de couleurs et style aux courbes

- `linestyle = '--'` Pour changer le style de la ligne, ici en pointiller  
- `color = 'purple'` choisir la couleur, ce référer aux "HTML Color Names" pour trouver les noms des différentes couleurs   
- `marker = 's'` Poser des marqueurs sur chaques points de la courbe
- On peut également utliser des raccourcis, comme passer juste comme 3ème argument après `plt.plot(x, y, .ici.)` simplement une lettre pour la couleur et un sigle pour la marque du point, et le type de ligne, `'b+--` == courbe bleu, points en croix, traceé en tirets

```python
plt.plot(mois, depenses, linestyle='--', color='purple')#(x |, y –) 
plt.plot(mois, revenue, color='#82edc9', marker='s')
plt.show()
```

### Zoomer avec `plt.axis()`

 Il peut-être utile de zoomer sur certaines parties d'un graphiques  
On peut utiliser pour cela `plt.axis()`, en lui passant une `list` des `min` et `max` des `x`et des `y` à afficher

- `plt.axis([min(x), max(x), min(y), max(y)])`


```python
plt.plot(mois, depenses, linestyle='--', color='purple')#(x |, y –) 
plt.plot(mois, revenue, color='#82edc9', marker='s')

plt.axis([2, 6, 130, 320])
#plt.axis([min(x), max(x), min(y), max(y)])

plt.show()
```

### Titres, légendes et noms des labels

- `plt.xlabel('..')` Titre vertical de l'axe des ordonnées `y` sur la gauche du plot  
- `plt.ylabel('...')` titre horizotal de l'axe des absisses `x` en dessous du plot    
- `plt.title('..')` titre du graphique ce place en haut
- `plt.legend(['.., ..'])` affiche le noms des différentes courbes paseé dans une `list`, prend en argument `loc` qui va de 0 à 10 et détermine la position du panneau


```python
plt.plot(mois, depenses, 'bo:')#(x –, y |) 
plt.plot(mois, revenue, 'r+--')

plt.ylabel("entrées / sortie d'argent")
plt.xlabel("temps en mois")

plt.legend(['depenses', 'revenue'], loc=2)#liste des noms de courbes
plt.title("visualisation de rentabilité")

plt.show()
```

### subplots   
pour implémenter plusieurs graphiques à la suite on utilise la commande qui suit.  
On lui passe en paramètres, le nombre de lignes puis de colonnes, comme pour créer un tableau de plusieurs graph, puis la position du graph dans se tableau 

- `plt.subplot(row, col, position)`  lignes, colonnes, et position du plot là dedans
- `plt.subplot_adjust()` permet d'ajuster la distance entre les plot pou pas qu'ils ne ce superpose. Les arguments possible de *__subplot_adjust()__* :
    - `left` `=` par défaut à 0.125, augmente ou diminue l'espace sur la gauche
    - `right` par défaut à 0.9
    - `bottom` marge en bas du plot, par défaut à 0.1
    - `top` marge au-dessus du plot à 0.1
    - `wspace` espace horizontale entre 2 plot les uns au-dessus des autres, à 0.2
    - `hspace` espace verticale enre 2 graph adjacents, à 0.2 par défaut

```python
x = range(7)
droite = [0, 1, 2, 3, 4, 5, 6]
parabole = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

# 1
plt.subplot(2, 1, 1)  # 2 lignes, 1 colonnes, position 1
plt.plot(x, droite, 'r+:')

# 2
plt.subplot(2, 2, 3) # 2lignes, 2 colonnes, pos 3
plt.plot(x, parabole, 'o')

# 3
plt.subplot(2, 2, 4)
plt.plot(x, cubic, '.')

plt.subplots_adjust(wspace=0.35, bottom=0.2)

plt.show()
```

_*__Autres exemples__ avec 6 subplots:*_

`plt.subplot(2, 3, 4)`  
2 lignes, 3 colonnes, position 4
se plot sera donc le 1er sur la 2ème ligne de la 1ère colonne

```python
plt.subplot(2, 3, 1)
plt.plot(x, cubic, 'y*')
plt.title('I')

plt.subplot(2, 3, 2)
plt.plot(x, parabole, 'ro')
plt.title('II')

plt.subplot(2, 3, 3)
plt.plot(x, droite, 'bx:')
plt.title('III')

plt.subplot(2, 3, 4)
plt.plot(x, parabole, 'go--')
plt.title('IV')

plt.subplot(2, 3, 5)
plt.plot(x, cubic, '--')
plt.title('V')

plt.subplot(2, 3, 6)
plt.plot(x, droite, '-.')
plt.title('VI')

plt.subplots_adjust(wspace=0.50, bottom=0.01, hspace=0.5, right=1.5)
plt.show()
```

### Changer l'échelles des axes 

Comment changer les noms et les valeurs des échelles des axes des abcisses et des ordonnées  
En utilisant l'objet `ax` on pourra modifier les l'échelles des axes d'un plot ou d'un subplot en particulier.

- `ax = plt.subplot()` on instancie l'objet `ax` avec un plot, on rajoute en arguments la position du subplot si il en a plusieurs  
- `ax.set_xticks()` échelle de valeurs chiffré pour l'axe des abcisses `x`, en argument une `list` de nb
- `ax.set_xticklabels()` étiquettes afficher pour représenter l'échelle ci-dessus, argument une `list` de de `str` représentant les valeurs réel de l'échelle.  
- `ax.set_yticks()` échelle de valeurs chiffré pour l'axe des ordonnées `y`
- `ax.set_yticklabels()` étiquettes afficher pour représenter les `y`

*__Exemple__*
- `ax.set_xticks()` = une `list` des chiffres de 1 à 12 (pour des mois)  
- `ax.set_xticklabels()` = une `list` de `str` des noms des mois ['jan',..] qui seront afficher
        

```python
plt.close('all')
```

```python
douze_mois = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]
nb_mois = range(12)
taux = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]
```

```python
plt.figure(figsize=(16, 4))

plt.xlabel("Mois")
plt.ylabel("Taux")

plt.plot(nb_mois, taux, 'b+-.') # On trace notre graph normalement

ax=plt.subplot() # instancie ax avec les propriété d'un subpolt

ax.set_xticks(nb_mois) # les chiffres pour l'échelle en abcisses
ax.set_xticklabels(douze_mois) # auxquels on associe ces étiquettes, ici le nom des 12 mois

ax.set_yticks([0.10, 0.25, 0.5, 0.75]) # les chiffres pour l'échelle en ordonnées
ax.set_yticklabels(["10%", "25%", "50%", "75%"]) # on leur associes les pourcentage correspondant qui seront affichés

plt.show()
```
