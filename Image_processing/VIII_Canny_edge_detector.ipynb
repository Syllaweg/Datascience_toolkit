# *Filtre de Canny pour trouver les Edges*

Le filtre de Canny est l'un des filtres les plus populaires pour traitement d'images, son algorithme est plus rapide et précis que celui de *Sobel*

Il va nous aider à détecter les formes et les bords des objets présents dans une image, afin par exemple de la diviser ...

L'avantage de représenter une image, avec une quantitée de données très largement réduite, mais en conservant grace au *Edges* des objets l'essentiels de l'information

```python
import numpy as np
import matplotlib.pyplot as plt
from helpers import show_image
```

```python
img_panneau = plt.imread('data/panneau_droit.png')
```

## Canny Edges Detector

- Il faut d'abord transformer l'image en 2D *Grayscale*

```python
from skimage.color import rgb2gray
```

```python
gray_panneau = rgb2gray(img_panneau)

show_image(gray_panneau)
```

### Filtre de Canny 
Paramètres: 
- *`sigma=`* règle l'intensité du filtre *Gaussien* qui va régler la sensibilité de la detection de bords. Plus *sigma* est petit, plus l'effet du filtre *Gaussien* appliqué sur l'image sera petit, et plus de bords et d'angles seront détecter avec précisions.  
Dans l'autre sens plus *`sigma`* est elevé plus le filtre *Guassien* retira de bruit, et les bords détectés seront moins saillant et précis..


```python
from skimage.feature import canny
```

```python
canny_panneau = canny(gray_panneau)
```

```python
show_image(canny_panneau)
```
