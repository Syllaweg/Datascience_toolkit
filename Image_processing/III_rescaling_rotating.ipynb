# *III Rotate & Rescale*

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
img = plt.imread('data/panneau_droit.png')
```

## tourner une image et remettre à l'échelle une image
### Avec l'anti_aliasing
empèche les effets de crénélage (pixel apparent)

```python
from skimage.transform import rotate, rescale
```

```python
# Rotate the image 90 degrees clockwise 
rotated_image = rotate(img, -90)

# Rescale with anti aliasing
rescaled_with_antialiasing = rescale(rotated_image, 1/4, anti_aliasing=True, multichannel=True)
```

```python
show_image(rescaled_with_antialiasing, "avec anti_aliasing")
```

### Sans anti_aliasing

```python
# Rescale without anti aliasing
rescaled_without_antialiasing = rescale(rotated_image, 1/4, anti_aliasing=False, multichannel=True )

# Show the resulting images
show_image(rescaled_with_antialiasing, "Transformé Avec anti aliasing")
show_image(rescaled_without_antialiasing, "Transformé Sans anti aliasing")
```

On observe bien que les contours des objet sont beaucoup plus crénelé de pixels sans ant_aliasing

## Agrandir une image
Avec l'algo de Scikit image on peut agrandir une image jusqu'a 3x, en réussissant à limiter les effets de floues

```python
from skimage.transform import rescale
```

```python
# Enlarge the image so it is 3 times bigger
enlarged_image = rescale(img, 3, anti_aliasing=True, multichannel=True)

# Show original and resulting image
show_image(img)
show_image(enlarged_image, "image 3x plus grande")
```

## redéfinir la taille d'une image proportionellement 

```python
pmz_img = plt.imread("data/pmz1.jpg")
```

```python
from skimage.transform import resize
```

```python
# divise la taille de l'image par 2, en recupérant les shape grace a numpy
height = int(pmz_img.shape[0]/ 2)
width = int(pmz_img.shape[1]/ 2)
```

```python
# Resize using the calculated proportional height and width
resized_pmz = resize(pmz_img, (height, width),
                       anti_aliasing=True)
```

```python
# Show the original and rotated image
show_image(pmz_img, 'Original')
show_image(resized_pmz, 'Resized image')
```
