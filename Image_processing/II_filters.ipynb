# *II Filters*

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
img = plt.imread('data/panneau_droit.png')
```


## Les Filtres
### Filtre de Sobel

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from skimage import color # pour utiliser color.rgb2gray et passer l'img en noir et blac d'abord
from skimage.filters import sobel
```

```python
# Make the image grayscale
img_gray = color.rgb2gray(img)

# Apply edge detection filter
edge_sobel = sobel(img_gray)

# Show original and resulting image to compare
show_image(img, "Original")
show_image(edge_sobel, "Edges with Sobel")
```

## Amliration du contraste

Dans les cas ou le contraste serait mauvais, cela nous permettera de faire apparaitre ou de mieux voire certaines choses

```python
from skimage import exposure
```

```python
# image original
show_image(img)
```

### *`equalize_hist()`*

- affiche l'histogramme de l'image

```python
plt.title('Histogram of image')
plt.hist(img.ravel(), bins=256)
plt.show()
```

- *`equalize_hist()`* va amliorer le contraste des pixels entre eux par rappport  l'histogramme de l'image et le rend donc plus uniforme

```python
# Use histogram equalization to improve the contrast
image_eq =  exposure.equalize_hist(img)

# Show the resulting image
show_image(image_eq, 'Resulting image')
```

On observe ici que l'ombre  disparue et que les bouchons sont plus visibles

### *`equalize_adapthist()`*
Pour une galisation adaptative des contraste en fonction des zones et touours de l'histogramme de l'image


```python
from skimage import exposure
```

```python
# Apply the adaptive equalization on the original image

adapthist_eq_image = exposure.equalize_adapthist(img, clip_limit=0.03)
```

```python
# Compare the original image to the equalized
show_image(img)
show_image(adapthist_eq_image)
```

On observe sans problme la disparition des zones d'ombres et une amlioration de la lisibit de l'image..
