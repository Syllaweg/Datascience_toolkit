# *Morphologie, éroder ou dilater un image*
On peut modifier la morphologir d'objet dans une image afin de mieux les dectecter ou de les restaurer

Ces opérations de dilataion ou érodage son conçue pour des *images binaires*, mais peuvent s'appliquer à des images en échelles de gris (grayscales)

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from helpers import show_image
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
img = plt.imread('data/panneau_droit.png')
```

```python
pmz_img = plt.imread('data/pmz1.jpg')
```

***

- Avant cela il faut d'abord transformer notre image en image binaire uniquement noir et blanc

```python
from skimage.filters import threshold_minimum, threshold_otsu
from skimage import color
```

```python
# Turn image to grayscale
gray_img = color.rgb2gray(img)

# Obtain the optimal threshold value
thresh = threshold_otsu(gray_img)

# Apply thresholding to the image
binary_img = gray_img > thresh
```

```python
plt.imshow(binary_img)
```

```python
skimage.filters.thresholding.threshold_minimum
```

`binary_img`est maintenant notre image en binaire 
***

## Éroder une image 

```python
from skimage import morphology
from skimage import data
```

```python
# Obtain the eroded shape 
eroded_image_shape = morphology.binary_erosion(binary_img)
```

```python
# See results
show_image(binary_img, 'Original')
show_image(eroded_image_shape, 'Eroded image')
```

On constate que les éléments de l'image on été érodés, les files par exemple sont plus fins.

## Dilater une image
Grossir les traits des objet de l'image..

```python
from skimage import morphology
```

```python
# Obtain the dilated image 
dilated_image = morphology.binary_dilation(binary_img)
```

```python
# See results
show_image(binary_img, 'Original')
show_image(dilated_image, 'Dilated image')
```

On peut voire que les file sont plus épais ...
